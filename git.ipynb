### Git Branches
- a branch represents an independent line of development
- the git branch command lets you create, list, rename and delete branches
- the default branch name in Git is master


- git branch --> to see current branch
- git branch branch-name  --> to add new branch
- git checkout branch-name --> to switch branches
- git checkout -b branch-name --> to create and switch at a time
- git branch -m old new --> to rename a branch 
- git clone -b branch-name repo-url  --> to clone specific branch
- git branch -d branch  --> to delete a branch

- The -d option will delete the branch only if it has already been pushed and merged with the remote branch. use -D instead if you want to force the branch to be deleted even if it hasn't been pushed ormerged yet. The branch deleted locally
