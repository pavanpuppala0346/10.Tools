## 1. How to install Git?
1. To install Git click on this link [https://git-scm.com/downloads](https://git-scm.com/downloads) Download as per your OS and install.
2. After installing Git open Git Bash.
3. Create a New folder and enter into that dir.      
4. First we need to initilize the git repository
    - <b>git status</b>
    - <b>git init</b>  --> to initialize the git repo(.git will create)
    - <b>ls</b>        --> to list the files 
    - <b>ls –a</b>     --> to list the hidden files 
5. by default branch is master 
6. After that we need to configure user name and email id   
    - <b>git config --global user.name "pavan"</b>
    - <b>git config --global user.email "pkumar.datascience@gmail.com"</b>
    - <b>git config --list</b> ( here we can list all configurations )
7. Create some sample files by using Touch or Vi or Cat commands
EX: Touch file1
vi file2
cat > file3
    - <b>touch file1</b>        --> to create empty file
    - <b>git add filename</b>   --> to promote code from workspace to staging area
    - <b>git status</b>         --> to check the files
    - <b>git commit –m”first commit”</b> -->to commit the code from stagging to local repo
    - <b>git log</b>            -->to check files in local repo

#### 2. log commands
- <b>git log</b>               --> it contains commit_id, author, date, commit_msge`m
- <b>git log --oneline</b>     -->    each commit_id with 7 characters and label message 
- <b>git log -n</b>            -->    "git log -2"latest two logs
- <b>git log --oneline -3</b>  -->    three latest commit_id’s with 7 characters 

- <b>git log --author=pavan</b>      -->       to check the logs by author name wise
- <b>git log --author=pavan -n</b>   -->        two latest logs by author name wise            
- <b>git log --since=yyyy-mm-dd</b>   
- <b>git log  --until=yyyy-mm-dd</b>
- <b>git log --oneline –decorate</b>  -->      it returns logs and with branch name
