#### 8. What are DBT Models
- SQL queries that define how raw data is transformed into more useful, structured data within the warehouse. 
- DBT model is a SELECT statement. 
- Models are .sql statements that contains transformation logic.

```python
from IPython.display import Image
Image('.\img\dbt13.png')
```

Every model in DBT is a Table. how can we replace instead of table to view using jinja template?
- if you not provide the jinja template on top of the dbt model then dbt looks for configure file by default it will create a table

for reference we will see the default model in DBT
- on top of the model we can specify whether model is table or view using jinja
    - {{ config(materialized='table')}} --> table
    - {{ config(materialized='view')}} --> view
- inside configuration yaml file dbt_project.yaml
    - models: example: +materialized : table
 
<b>Note: </b> First priority is jinja template, if jinja expression is not available then it look for configuration yaml file
